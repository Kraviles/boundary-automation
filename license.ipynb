```python
import requests

iso = "TZA"
release = "gbOpen"
adm = "ADM1"

url = f"https://www.geoboundaries.org/api/current/{release}/{iso}/{adm}/"

response = requests.get(url)
response.raise_for_status()

data = response.json()
print(data.keys())  # structure
```
